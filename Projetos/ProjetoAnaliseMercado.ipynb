## Analise de Mercado Financeiro

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
%matplotlib inline
```

```python
dataInicio = dt.datetime(2008, 1, 1);
dataFim = dt.datetime(2018, 5, 29);
```

```python
bac = data.DataReader('BAC','morningstar',dataInicio,dataFim)
c = data.DataReader('C','morningstar',dataInicio,dataFim)
gs = data.DataReader('GS','morningstar',dataInicio,dataFim)
jpm = data.DataReader('JPM','morningstar',dataInicio,dataFim)
ms = data.DataReader('MS','morningstar',dataInicio,dataFim)
wfc = data.DataReader('WFC','morningstar',dataInicio,dataFim)

#falta resetar o index aqui.... colocar data 
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
bancos = pd.concat([bac, c, gs, jpm, ms, wfc], axis=1, keys=tickers)
bancos.columns.names = ['Bank Ticker','Stock Info']
bancos
```

```python
#Qual é o preço máximo de fechamento para o estoque de cada banco durante todo o período? 
dafFechamamento = bancos.xs(key='Close',axis=1,level='Stock Info') #coluna que ele vai mostrar, eixo 1 referente a coluna, ja que é multnivel o segundo nivel esta a coluna Stock
dafFechamamento.plot()
dafFechamamento.max()
```

```python
returns = pd.DataFrame()
```

** Crie um novo DataFrame vazio chamado returns. Este dataframe conterá os retornos para o ação de cada banco. Os retornos geralmente são definidos por: **

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
for tick in tickers:
    returns[tick+' Return'] = bancos[tick]['Close'].pct_change()
    
returns.head()
```

## Médias móveis

Vamos analisar as médias móveis para essas ações no ano de 2008.

** Trace a média de 30 dias para o preço próximo do Bank Of America para o ano de 2008 **

```python
plt.figure(figsize=(12,6))
bac['Close'].ix['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
bac['Close'].ix['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
plt.legend()
```

```python
#Crie um mapa de calor da correlação entre os preços de fechamento das ações. 
sns.heatmap(bancos.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

```python
sns.clustermap(bancos.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```
