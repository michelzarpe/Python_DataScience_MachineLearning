**Distribuição dos plots**
--
    *Distplot, jointplot, pairplot, rugplot, kdeplot

```python
import seaborn as sns
import pandas as nd
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') #com isso retorna um dataFrame, de tips, gorjeta 
```

```python
tips.info()
```

```python
tips.head()
#total_conta valor_gorjeta, sexo, fumante, dia,tempo, quantidade
```

```python
#DistPlot
#Em estatística, estimativa de densidade Kernel (EDK) é uma forma não-paramétrica para estimar a
#função de densidade de probabilidade de uma variável aleatória. Estimação da densidade Kernel é um problema fundamental de 
#suavização de dados onde inferências sobre a população são feitas, com base numa amostra de dados finita.
#faz um KDE, FORMA DE DISTRIBUICAO não PARAmétrica, sem ter certeza a forma de distribuicao
sns.distplot(tips['total_bill'])
sns.rugplot(tips['total_bill']);
```

```python
#utilizado para grafico de distribuição conjunta
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
#pega todas as variaveis numericas e faz joinplots dela
sns.pairplot(tips);
```

```python
#pega todas as variaveis numericas e faz joinplots dela
#hue='' saber separado por categoria de sexo
sns.pairplot(tips,hue='sex');
```

```python
import seaborn as sns
import pandas as nd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
iris['species'].value_counts()
```

```python
grafico = sns.PairGrid(iris)
grafico.map(plt.scatter) #graficos de disperção
```

```python
grafico = sns.PairGrid(iris)
grafico.map_diag(plt.hist) #graficos de histogramas
grafico.map_upper(plt.scatter) #graficos de disperção
grafico.map_lower(sns.kdeplot) #graficos KDE
```

```python
sns.pairplot(iris,hue='species')
```
