```python
#Operações com Series 
paisesCopa = ['Brasil','Argentina','Portugal','Alemanha',"Equador"];
habilidade = [9.8,9.4,9.0,9.8,5.9];
potencialDosPaises = pd.Series(habilidade,paisesCopa)
print(potencialDosPaises)
```

```python
paisesForaCopa=["Italia","Venezuela"];
habilidade = [8,5.6];
potencialDosPaisesForaCopa = pd.Series(habilidade,paisesForaCopa)

paisesCopaRepescagem=["Equador"];
habilidade = [2];
potencialPaisesCopaRepescagem = pd.Series(habilidade,paisesCopaRepescagem)
```

```python
print(potencialDosPaises)
print("-----//----------")
print(potencialDosPaisesForaCopa)
print("-------//--------")
print(potencialPaisesCopaRepescagem)
```

```python
print(potencialDosPaises+potencialPaisesCopaRepescagem)
print(potencialDosPaises+potencialPaisesCopaRepescagem+potencialDosPaisesForaCopa)
```
